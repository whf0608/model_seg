```python
   


    for epoch in range(1, epochs+1):
        model.train()
        epoch_loss = 0
        with tqdm(total=100, desc=f'Epoch {epoch}/{epochs}', unit='img') as pbar:
            losss=[]
            for images in train_loader:
                
                t2 = images['t2'].to(device)  # .half()
                t2_b = images['t2_b'].to(device, dtype=torch.long)  # .half()
                mask2_b = images['mask2_b'].to(device, dtype=torch.long)  # .half()
                mask2 = images['mask2'].to(device, dtype=torch.long)  # .half()
                
                t2 = images['t2'].to(device)#.half()
                mask1_b = images['mask1_b'].to(device,dtype=torch.long)#.half()
                mask1 = images['mask1'].to(device,dtype=torch.long)#.half()
                
                
                with torch.cuda.amp.autocast(enabled=amp):
                    r1,r2,r3,r4,r5 = model(t2)
                    loss = criterion(r1,t2_b)\
                           + dice_loss(F.softmax(r1, dim=1).float(),
                                       F.one_hot(t2_b.squeeze(1), 3).permute(0, 3, 1, 2).float(),
                                       multiclass=True)\
                           + criterion(r2, mask2_b)\
                           + dice_loss(F.softmax(r2, dim=1).float(),
                                       F.one_hot(mask2_b.squeeze(1), 3).permute(0, 3, 1, 2).float(),
                                       multiclass=True)\
                            + criterion(r3, mask2)\
                           + dice_loss(F.softmax(r3, dim=1).float(),
                                       F.one_hot(mask2.squeeze(1), 3).permute(0, 3, 1, 2).float(),
                                       multiclass=True)\
                            + criterion(r4, mask1_b)\
                           + dice_loss(F.softmax(r4, dim=1).float(),
                                       F.one_hot(mask1_b.squeeze(1), 3).permute(0, 3, 1, 2).float(),
                                       multiclass=True)\
                            + criterion(r5, mask1)\
                           + dice_loss(F.softmax(r5, dim=1).float(),
                                       F.one_hot(mask1.squeeze(1), 3).permute(0, 3, 1, 2).float(),
                                       multiclass=True)
                    loss/=5.0
                optimizer.zero_grad(set_to_none=True)
                grad_scaler.scale(loss).backward()
                grad_scaler.step(optimizer)
                grad_scaler.update()
                losss.append(loss.item())
                pbar.update(t2.shape[0])
                pbar.set_postfix(**{'loss (batch)': loss.item()})
            pbar.set_postfix(**{'loss (epoch)': sum(losss)/len(losss)})  
            
    torch.save(model,'test15.pt')
```
