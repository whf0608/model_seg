```python
def update_paramter(model, epoch):
    learning_rate: float = 1e-5
    learning_rate1: float = 1e-7
    
    if epoch>=100:
         learning_rate= learning_rate-epoch*1e-7*0.5
    
    optimizer = [{ 'lr':learning_rate1, 'weight_decay':1e-8,'momentum': 0.9}]
    optimizer.append({ 'lr':learning_rate1, 'weight_decay':1e-8,'momentum': 0.9})
    optimizer.append({'lr':learning_rate1, 'weight_decay':1e-8,'momentum': 0.9})
    optimizer.append({'lr':learning_rate1,'weight_decay':1e-8,'momentum': 0.9})
    optimizer.append({ 'lr':learning_rate1,'weight_decay':1e-8,'momentum': 0.9})
    optimizer.append({ 'lr':learning_rate1,'weight_decay':1e-8,'momentum': 0.9})
    if epoch<100:
        
        optimizer[int(1+epoch/5%5)]['lr']=learning_rate-epoch%5*0.2*learning_rate
        
        learning_rate= learning_rate-epoch*1e-7*0.5
        optimizer[0]['lr']=learning_rate-epoch%5*0.2*learning_rate
    else:
        optimizer[0]['lr']=learning_rate-epoch%5*0.2*learning_rate
        for i in range(1,6):
            optimizer[i]['lr']=learning_rate -epoch%5*0.2*learning_rate
        
 
    return optimizer
```

```python
lrs_a=[[],[],[],[],[],[]]
for epoch  in range(200):
    lrs = update_paramter(model=None, epoch=epoch)
    for i,lr in enumerate(lrs):
        lrs_a[i].append(lr['lr'])
```
